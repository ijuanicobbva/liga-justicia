```scala
import Constants._
import org.apache.spark.sql.functions.{col, trim, upper, regexp_replace, lit, when, coalesce}
import org.apache.spark.sql.{DataFrame, SparkSession}
import org.apache.commons.lang3.exception.ExceptionUtils
import org.apache.spark.sql.types.StringType


class InputReader(spark: SparkSession, hdate: String, odate: String) {
  
  def readGuaranteeCore(path: String): DataFrame = {
    try {
      spark.read.parquet(path + cutoff_date_partition + odate)
        .select(
          trim(col(contract_id)).as(contract_id),
          trim(col(guarantee_type)).as(guarantee_type),
          trim(col(isin_id)).as(isin_id_coll),
          col(current_guaranteed_amount)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def readRentaFijaOpeBis(path: String): DataFrame = {
    try {
      spark.read.parquet(path + load_date_partition + hdate)
        .filter(!trim(col(market_operation_type)).isin(prods:_*))
        .select(
          trim(col(customer_platform_id)).as(customer_platform_id),
          trim(col(customer_corporate_id)).as(customer_corporate_id),
          when(col(operation_deal_front_star_id).isNull, lit(emptyString))
            .otherwise(regexp_replace(col(operation_deal_front_star_id), blank_spaces,emptyString)).as(operation_deal_front_star_id),
          when(col(source_application_id).isNull, lit(emptyString))
            .otherwise(regexp_replace(col(source_application_id), blank_spaces, emptyString)).as(source_application_id),
          when(col(operation_deal_back_star_id).isNull, lit(emptyString))
            .otherwise(regexp_replace(col(operation_deal_back_star_id), blank_spaces,emptyString)).as(operation_deal_back_star_id),
          col(total_securities_nominal_amount),
          col(market_operation_start_date),
          trim(col(coupon_rate_type)).as(coupon_rate_type),
          col(coupon_term_days_number),
          col(return_interest_rate_per),
          col(current_rate_coupon_per),
          trim(col(coupon_rate_name)).as(coupon_rate_name),
          col(market_operation_settlement_date),
          col(current_commitment_amount),
          col(initial_amount),
          col(seller_branch_id),
          trim(col(market_operation_type)).as(market_operation_type),
          trim(col(isin_id)).as(isin_id),
          trim(col(contract_id)).as(contract_id)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def readProductsIntRateOptions(path: String): DataFrame = {
    try {
      spark.read.parquet(path + cutoff_date_partition + hdate)
        .select(
          when(col(oper_origin_app_assigned_id).isNull, lit(emptyString))
            .otherwise(regexp_replace(col(oper_origin_app_assigned_id), blank_spaces,emptyString)).as(operation_deal_front_star_id),
          trim(col(app_counterparty_id)).as(counterpart_id),
          trim(col(coupon_rate_name)).as(count_convention),
          trim(col(portfolio_investment_type)).as(portfolio_investment_type),
          col(operation_parts_number),
          col(operation_stage_number),
          col(operation_sire_assigned_id),
          col(purchase_sale_ctrct_amount),
          col(strike_amount),
          coalesce(col(coupon_rate_name), col(prch_sale_option_ir_cvtn_type)).as(coupon_rate_name),
          col(fin_derv_oper_phase_end_date),
          col(fin_derv_oper_phase_start_date)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def readForwardIntRateProducts(path: String): DataFrame = {
    try {
      spark.read.parquet(path + cutoff_date_partition + hdate)
        .select(
          when(col(oper_origin_app_assigned_id).isNull, lit(emptyString))
            .otherwise(regexp_replace(col(oper_origin_app_assigned_id), blank_spaces,emptyString)).as(operation_deal_front_star_id),
          trim(col(app_counterparty_id)).as(counterpart_id),
          trim(col(calculation_interest_name)).as(count_convention),
          trim(col(portfolio_investment_name)).as(portfolio_investment_name),
          col(operation_parts_number),
          col(operation_stage_number),
          trim(col(pymt_rcpt_fwd_oper_mark_type)).as(pymt_rcpt_fwd_oper_mark_type),
          col(operation_sire_assigned_id),
          col(nominal_amount),
          trim(col(fndr_variable_rate_int_type)).as(fndr_variable_rate_int_type),
          trim(col(calculation_interest_name)).as(coupon_rate_name),
          col(fin_derv_oper_phase_end_date),
          col(fin_derv_oper_phase_start_date),
          col(underlying_expiry_date)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def dataCountConvention(productsIntRateOptions: DataFrame, forwardIntRateProducts: DataFrame, tsiSwapsTasa: DataFrame,
                          tsiSwapsIv: DataFrame, tsiSwapsCre: DataFrame): DataFrame = {
    try {
      productsIntRateOptions.select(
        col(operation_deal_front_star_id),
        col(counterpart_id),
        col(count_convention)
      )
        .union(forwardIntRateProducts.select(
          col(operation_deal_front_star_id),
          col(counterpart_id),
          col(count_convention)
        )
        )
        .union(
          tsiSwapsTasa.select(
            col(operation_deal_front_star_id),
            col(counterpart_id),
            col(count_convention)
          )
        )
        .union(tsiSwapsIv.select(
          col(operation_deal_front_star_id),
          col(counterpart_id),
          col(count_convention)
        )
        )
        .union(tsiSwapsCre.select(
          col(operation_deal_front_star_id),
          col(counterpart_id),
          col(count_convention)
        )
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def readSireFxPrdt(path: String): DataFrame = {
    try {
      spark.read.parquet(path + load_date_partition + hdate)
        .filter(col(subproduct_type) =!= FXLI)
        .select(
          col(counterparty_id).as(counterpart_id),
          when(col(source_application_id).isNull, lit(emptyString))
            .otherwise(regexp_replace(col(source_application_id), blank_spaces,emptyString)).as(source_application_id),
          trim(col(operation_deal_front_star_id)).as(operation_deal_front_star_id),
          trim(col(operation_deal_back_star_id)).as(operation_deal_back_star_id),
          col(market_oper_recv_notional_amount),
          col(forward_price_amount),
          col(seller_branch_id),
          trim(col(operation_nature_type)).as(operation_nature_type)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def readDerivadosMmooBis(path: String): DataFrame = {
    try {
      spark.read.parquet(path + load_date_partition + hdate)
        .select(
          col(counterparty_systems_oper_id).as(counterpart_id),
          when(trim(col(warrant_type)).isin(C, CALL), CALL).when(trim(col(warrant_type)).isin(P, PUT), PUT).otherwise(lit(X)).as(warrant_type),
          trim(col(product_type)).as(product_type),
          when(col(source_application_id).isNull, lit(emptyString))
            .otherwise(regexp_replace(col(source_application_id), blank_spaces,emptyString)).as(source_application_id),
          trim(col(operation_deal_front_star_id)).as(operation_deal_front_star_id),
          trim(col(operation_deal_back_star_id)).as(operation_deal_back_star_id),
          trim(col(trading_mark_type)).as(trading_mark_type),
          col(market_oper_pay_notional_amount),
          trim(col(option_style_mark_type)).as(option_style_mark_type),
          col(strike_amount),
          trim(col(underline_reference_type)).as(underline_reference_type),
          col(contract_amount),
          col(expiration_date),
          col(market_oper_recv_notional_amount),
          col(value_date),
          col(forward_swap_recv_rate_amount),
          col(forward_swap_pay_rate_amount),
          col(isin_id),
          trim(col(seller_branch_id)).as(seller_branch_id)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def readFinDervOtcBis(path: String): DataFrame = {
    try {
      spark.read.parquet(path + load_date_partition + hdate)
        .filter(trim(col(subproduct_type)) =!= CAP)
        .filter(!(trim(col(subproduct_type)).isin(IRC, IRS) && trim(col(gl_account_id)).substr(0,1).isin("1","2") ||
          trim(col(gl_account_type)) === H))
        .select(
          trim(col(securities_dealing_desk_name)).as(portfolio_type),
          trim(col(counterparty_id)).as(counterpart_id),
          when(trim(col(warrant_type)).isin(C, CALL), CALL).when(trim(col(warrant_type)).isin(P, PUT), PUT).as(warrant_type),
          trim(col(operation_deal_front_star_id)).as(operation_deal_front_star_id),
          when(col(source_application_id).isNull, lit(emptyString))
            .otherwise(regexp_replace(col(source_application_id), blank_spaces,emptyString)).as(source_application_id),
          trim(col(operation_deal_back_star_id)).as(operation_deal_back_star_id),
          trim(col(option_style_mark_type)).as(option_style_mark_type),
          trim(col(product_type)).as(product_type),
          col(expiration_date),
          col(forward_swap_pay_rate_amount),
          col(forward_swap_recv_rate_amount),
          trim(col(subproduct_type)).as(subproduct_type),
          trim(col(underline_reference_type)).as(underline_reference_type),
          col(value_date),
          trim(col(trading_mark_type)).as(trading_mark_type),
          trim(col(isin_id)).as(isin_id),
          trim(col(seller_branch_id)).as(seller_branch_id)
        )
        .distinct()
    } catch {
      case e: Exception=> ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def readInternalOperMurex(path: String): DataFrame = {
    try {
      spark.read.parquet(path + load_date_partition + hdate)
        .select(
          trim(col(front_system_number_id)).as(operation_deal_front_star_id),
          trim(col(portfolio_investment_name)).as(portfolio_investment_name),
          trim(col(app_counterparty_id)).as(counterpart_id),
          col(operation_parts_number),
          trim(col(fin_derivative_leg_mark_type)).as(fin_derivative_leg_mark_type),
          trim(col(fin_derv_leg_exch_int_mark_type)).as(fin_derv_leg_exch_int_mark_type)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def readLiquidityFlowsOverSwap(path: String): DataFrame = {
    try {
      spark.read.parquet(path + cutoff_date_partition + hdate)
        .filter(col(fin_derv_oper_curr_flow_type) === FL && lit(hdate).between(col(fin_derv_flow_start_date), col(period_end_date)))
        .select(
          trim(col(murex_operation_id)).as(operation_deal_front_star_id),
          trim(col(operation_sire_assigned_id)).as(sire_id),
          col(forward_price_amount),
          trim(col(oper_flow_pymt_rcpt_mark_type)).as(oper_flow_pymt_rcpt_mark_type)
        )
        .distinct()
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def readLiqOverRateOptions(path: String): DataFrame = {
    try {
      spark.read.parquet(path + cutoff_date_partition + hdate)
        .filter(col(fin_derv_oper_curr_flow_type) === FL && lit(hdate).between(col(fin_derv_flow_start_date), col(underlying_expiry_date)))
        .select(
          trim(col(operation_sire_assigned_id)).as(sire_id),
          trim(col(oper_origin_app_assigned_id)).as(operation_deal_front_star_id),
          col(forward_price_amount)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def readLiqOverRateOptionsOpt(path: String): DataFrame = {
    try {
      spark.read.parquet(path + cutoff_date_partition + hdate)
        .filter(col(fin_derv_oper_curr_flow_type) === FL)
        .select(
          trim(col(operation_sire_assigned_id)).as(sire_id),
          trim(col(oper_origin_app_assigned_id)).as(operation_deal_front_star_id),
          col(fixed_var_flow_leg_mark_type)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def subproductCfc(tsiOperacion: DataFrame, liqOverRateOptionsOpt: DataFrame): DataFrame = {
    try {
      tsiOperacion.join(liqOverRateOptionsOpt, Seq(operation_deal_front_star_id, sire_id), inner)
        .filter(col(financial_product_id) === CFC)
        .drop(operation_sire_assigned_id)
        .select(
          trim(col(operation_deal_front_star_id)).as(operation_deal_front_star_id),
          col(fixed_var_flow_leg_mark_type)
        )
        .distinct()
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def tsiEconomicData(tsiSwapsIv: DataFrame, tsiSwapsCre: DataFrame, tsiSwapsTasa: DataFrame, productsIntRateOptions: DataFrame,
                      forwardIntRateProducts: DataFrame): DataFrame = {
    try {
      tsiSwapsIv.select(
        col(operation_sire_assigned_id), col(operation_deal_front_star_id), col(operation_parts_number), col(operation_stage_number),
        col(nominal_amount), lit(emptyString).as(strike_amount), lit(emptyString).as(fndr_variable_rate_int_type), col(count_convention),
        col(fin_derv_leg_flow_frequency_name), lit(emptyString).as(break_clause_frequency_type), col(operation_repayment_number),
        col(fin_derv_oper_phase_end_date), col(fin_derv_active_value_start_date), lit(emptyString).as(fin_derv_oper_phase_start_date),
        lit(emptyString).as(underlying_expiry_date), col(fin_derv_leg_exch_int_mark_type), col(operations_date_payment_type),
        lit(emptyString).as(fin_derivative_settlement_amount), col(fin_derivative_leg_mark_type), col(current_spread_amount), col(interest_rate_amount)
      ).union(
          tsiSwapsCre.select(
            col(operation_sire_assigned_id), col(operation_deal_front_star_id), col(operation_parts_number), col(operation_stage_number),
            col(nominal_amount), lit(emptyString), lit(emptyString), col(count_convention), lit(emptyString), lit(emptyString),
            lit(emptyString), col(fin_derv_oper_phase_end_date), col(fin_derv_active_value_start_date), col(fin_derv_oper_phase_start_date),
            lit(emptyString), lit(emptyString), lit(emptyString), lit(emptyString), col(fin_derivative_leg_mark_type),
            col(current_spread_amount), lit(emptyString)
          )).union(
          tsiSwapsTasa.select(
            col(operation_sire_assigned_id), col(operation_deal_front_star_id), col(swap_legs_id), col(operation_stage_number),
            col(nominal_amount), lit(emptyString), lit(emptyString), col(count_convention), col(fin_derv_leg_flow_frequency_name),
            col(break_clause_frequency_type), col(operation_repayment_number), col(fin_derv_oper_phase_end_date), col(fin_derv_active_value_start_date),
            col(fin_derv_oper_phase_start_date), col(underlying_expiry_date), col(fin_derv_leg_exch_int_mark_type), col(operations_date_payment_type),
            col(fin_derivative_settlement_amount), lit(emptyString), col(current_spread_amount), col(interest_rate_amount)
          )).union(
          productsIntRateOptions.select(
            col(operation_sire_assigned_id), col(operation_deal_front_star_id), col(operation_parts_number), col(operation_stage_number),
            col(purchase_sale_ctrct_amount), col(strike_amount), lit(emptyString), col(coupon_rate_name), lit(emptyString), lit(emptyString),
            lit(emptyString), col(fin_derv_oper_phase_end_date), lit(emptyString), col(fin_derv_oper_phase_start_date), lit(emptyString),
            lit(emptyString), lit(emptyString), lit(emptyString), lit(emptyString), lit(emptyString), lit(emptyString)
          )).union(
          forwardIntRateProducts.select(
            col(operation_sire_assigned_id), col(operation_deal_front_star_id), col(operation_parts_number), col(operation_stage_number),
            col(nominal_amount), lit(emptyString), col(fndr_variable_rate_int_type), col(coupon_rate_name), lit(emptyString),
            lit(emptyString), lit(emptyString), col(fin_derv_oper_phase_end_date), lit(emptyString), col(fin_derv_oper_phase_start_date),
            col(underlying_expiry_date), lit(emptyString), lit(emptyString), lit(emptyString), lit(emptyString), lit(emptyString), lit(emptyString)
          )).select(
          col(operation_sire_assigned_id).as(sire_id), col(operation_deal_front_star_id), col(operation_parts_number), col(operation_stage_number),
          col(nominal_amount), col(strike_amount), col(fndr_variable_rate_int_type), col(count_convention).as(coupon_rate_name),
          col(fin_derv_leg_flow_frequency_name), col(break_clause_frequency_type), col(operation_repayment_number), col(fin_derv_oper_phase_end_date),
          col(fin_derv_active_value_start_date), col(fin_derv_oper_phase_start_date), col(underlying_expiry_date), col(fin_derv_leg_exch_int_mark_type),
          col(operations_date_payment_type), col(fin_derivative_settlement_amount), col(fin_derivative_leg_mark_type).as(mark_type),
          col(current_spread_amount), col(interest_rate_amount)
        ).distinct()
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def tsiEconomicDataSwaps(tsiSwapsDiv: DataFrame, internalOperMurex: DataFrame): DataFrame = {
    try {
      tsiSwapsDiv.select(lit(emptyString).as(sire_id), col(operation_deal_front_star_id), col(operation_parts_number), col(operation_stage_number),
        col(nominal_amount), lit(emptyString).as(strike_amount), lit(emptyString).as(fndr_variable_rate_int_type), col(coupon_rate_name),
        col(fin_derv_leg_flow_frequency_name), lit(emptyString).as(break_clause_frequency_type), col(operation_repayment_number),
        lit(emptyString).as(fin_derv_oper_phase_end_date), lit(emptyString).as(fin_derv_active_value_start_date),
        lit(emptyString).as(fin_derv_oper_phase_start_date), lit(emptyString).as(underlying_expiry_date), col(fin_derv_leg_exch_int_mark_type),
        col(operations_date_payment_type), lit(emptyString).as(fin_derivative_settlement_amount), col(payment_indicator_type),
        lit(emptyString).as(current_spread_amount), lit(emptyString).as(interest_rate_amount)
      ).union(internalOperMurex.select(
            lit(nullString).cast(StringType).as(sire_id), col(operation_deal_front_star_id), col(operation_parts_number), lit(emptyString),
            lit(emptyString), lit(emptyString), lit(emptyString), lit(emptyString), lit(emptyString), lit(emptyString), lit(emptyString),
            lit(emptyString), lit(emptyString), lit(emptyString), lit(emptyString), col(fin_derv_leg_exch_int_mark_type), lit(emptyString),
            lit(emptyString), col(fin_derivative_leg_mark_type), lit(emptyString), lit(emptyString))
        ).select(col(sire_id), col(operation_deal_front_star_id), col(operation_parts_number), col(operation_stage_number),
          col(nominal_amount), col(strike_amount), col(fndr_variable_rate_int_type), col(coupon_rate_name),
          col(fin_derv_leg_flow_frequency_name), col(break_clause_frequency_type), col(operation_repayment_number), col(fin_derv_oper_phase_end_date),
          col(fin_derv_active_value_start_date), col(fin_derv_oper_phase_start_date), col(underlying_expiry_date), col(fin_derv_leg_exch_int_mark_type),
          col(operations_date_payment_type), col(fin_derivative_settlement_amount), col(payment_indicator_type).as(mark_type), col(current_spread_amount),
          col(interest_rate_amount))
        .distinct()
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def tsiEconomicDataSwapsOpc(liqOverRateOptions: DataFrame, liquidityFlowsOverSwap: DataFrame): DataFrame = {
    try {
     liqOverRateOptions.union(
        liquidityFlowsOverSwap.select(
          col(operation_deal_front_star_id),
          col(sire_id),
          col(forward_price_amount)
        )
      )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def readliquidityFlowsOverSwapAux(path: String): DataFrame = {
    try {
      spark.read.parquet(path + cutoff_date_partition + hdate)
        .filter(trim(col(oper_flow_pymt_rcpt_mark_type)).isNotNull)
        .select(
          col(murex_operation_id),
          trim(col(operation_sire_assigned_id)).as(operation_sire_assigned_id),
          col(swap_legs_id),
          trim(col(oper_flow_pymt_rcpt_mark_type)).as(oper_flow_pymt_rcpt_mark_type)
        )
        .distinct()
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def activeOperationsEconomicData(tsiOperacion: DataFrame, tsiEconomicData: DataFrame, tsiEconomicDataSwapsOpc: DataFrame,
                                   liquidityFlowsOverSwapAux: DataFrame): DataFrame = {
    try {
      tsiOperacion.join(tsiEconomicData, Seq(operation_deal_front_star_id, sire_id), left)
        .join(tsiEconomicDataSwapsOpc, Seq(operation_deal_front_star_id, sire_id), left)
        .join(liquidityFlowsOverSwapAux, col(operation_deal_front_star_id) === col(murex_operation_id) &&
          col(operation_sire_assigned_id) === tsiOperacion.col(sire_id), left)
        .drop(sire_id, murex_operation_id, mark_type, operation_sire_assigned_id, oper_flow_pymt_rcpt_mark_type)
        .distinct()
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def activeOperationsEconomicDataSwaps(tsiOperacion: DataFrame, tsiEconomicDataSwaps: DataFrame, tsiOpcionDiv: DataFrame,
                                        liquidityFlowsOverSwapAux: DataFrame): DataFrame = {
    try {
      tsiOperacion.join(tsiEconomicDataSwaps, Seq(operation_deal_front_star_id), left)
        .join(tsiOpcionDiv, Seq(operation_deal_front_star_id), left)
        .join(liquidityFlowsOverSwapAux, col(operation_deal_front_star_id) === col(murex_operation_id) &&
          col(oper_flow_pymt_rcpt_mark_type) === col(mark_type), left)
        .drop(sire_id, murex_operation_id, mark_type, operation_sire_assigned_id, oper_flow_pymt_rcpt_mark_type)
        .distinct()
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def activeOperations(activeOperationsEconomicData: DataFrame, activeOperationsEconomicDataSwaps: DataFrame): DataFrame = {
    try {
      activeOperationsEconomicData.union(activeOperationsEconomicDataSwaps)
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def readTsiOperacionOtc(path:String): DataFrame = {
    try {
      spark.read.parquet(path + cutoff_date_partition + hdate)
        .select(
          trim(col(sire_oper_related_info_type)).as(sire_oper_related_info_type),
          trim(col(oper_origin_app_assigned_id)).as(operation_deal_front_star_id),
          trim(col(portfolio_investment_name)).as(portfolio_type),
          trim(col(app_counterparty_id)).as(counterpart_id),
          trim(col(market_operation_type)).as(market_operation_type)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def readTsiDepositos(path: String): DataFrame = {
    try {
      spark.read.parquet(path + cutoff_date_partition + hdate)
        .select(
          trim(col(oper_origin_app_assigned_id)).as(operation_deal_front_star_id),
          trim(col(operation_sire_assigned_id)).as(sire_id),
          col(operation_stage_number),
          trim(col(fin_derv_leg_flow_frequency_name)).as(fin_derv_leg_flow_frequency_name),
          trim(col(calculation_interest_name)).as(calculation_interest_name),
          trim(col(amortization1_id)).as(amortization1_id),
          trim(col(amortization_id)).as(amortization_id)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def readInternalOperMurexDepos(path: String): DataFrame = {
    try {
      spark.read.parquet(path + load_date_partition + hdate)
        .filter(trim(col(transaction_product_type)) === LoansDeposits)
        .select(
          col(front_system_number_id).as(operation_deal_front_star_id),
          col(current_operation_nom_amount),
          trim(col(transaction_type)).as(transaction_type),
          trim(col(fin_derv_leg_exch_int_mark_type)).as(fin_derv_leg_exch_int_mark_type),
          col(forward_rate_amount),
          trim(col(underlying_main_index_id)).as(underlying_main_index_id),
          col(forward_price_amount)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def readMdcoMoneyMktBis(path: String): DataFrame = {
    try {
      spark.read.parquet(path + load_date_partition + hdate)
      .select(
        trim(col(operation_deal_front_star_id)).as(operation_deal_front_star_id),
        trim(col(seller_branch_id)).as(seller_branch_id)
      )     
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e    
    }
  }      
}
```

```scala
import Constants._
import org.apache.spark.sql.functions.{abs, broadcast, coalesce, col, concat, current_timestamp, date_format, datediff, lit, lpad, months_between, regexp_replace, split, trim, udf, upper, when}
import org.apache.spark.sql.DataFrame
import org.apache.commons.lang3.exception.ExceptionUtils
import org.apache.spark.sql.types.{DateType, DecimalType, IntegerType, StringType}
import scala.util.Random

class Process(hdate: String) {

  val dateCalendar = udf((x: Int) =>
    try {
      if (x <= 365) {
        x. + ("D")
      }
      else if (x > 365 && (x / 30) > 100) {
        (x / 365). + ("Y")
      }
      else if (x > 365 && (x / 7) < 100) {
        (x / 7). + ("W")
      }
      else if (x > 365 && (x / 7) > 100) {
        (x / 30). + ("M")
      }
      else {
        x.toString
      }
    } catch {
      case e: Exception => lit("00000").toString
    }
  )

  def fixedRent(rentaFijaOpeBis: DataFrame, tsiContraparte: DataFrame, dataCountConvention: DataFrame,
                catalogApp: DataFrame, catalogPrsPcv: DataFrame, guaranteeCore: DataFrame): DataFrame = {
    try {
      rentaFijaOpeBis.join(tsiContraparte, Seq(customer_platform_id, customer_corporate_id), left)
        .join(dataCountConvention, Seq(operation_deal_front_star_id, counterpart_id), left)
        .join(catalogApp, rentaFijaOpeBis.col(source_application_id) === catalogApp.col(id), left)
        .join(broadcast(catalogPrsPcv), rentaFijaOpeBis.col(market_operation_type) === catalogPrsPcv.col(id), left)
        .join(guaranteeCore, Seq(contract_id), left)
        .select(coalesce(catalogApp.col(value), when(rentaFijaOpeBis.col(source_application_id) === emptyString, BGM_BIS)
          .otherwise(rentaFijaOpeBis.col(source_application_id))).as(gf_origin_application_id),
          coalesce(catalogApp.col(value), when(rentaFijaOpeBis.col(source_application_id) === emptyString, BGM_BIS)
            .otherwise(rentaFijaOpeBis.col(source_application_id))).as(gf_front_sys_name),
          when(col(operation_deal_front_star_id) === emptyString, lit(tenZero)).when(col(operation_deal_front_star_id) === zero_string, lit(tenZero))
            .when(col(operation_deal_front_star_id) === 0, col(operation_deal_back_star_id))
            .otherwise(col(operation_deal_front_star_id)).as(gf_front_sys_trd_id),
          when(col(operation_deal_back_star_id) === emptyString, lit(tenZero))
            .when(col(operation_deal_back_star_id) === zero_string, lit(tenZero)).otherwise(col(operation_deal_back_star_id)).as(gf_back_sys_trd_id),
          lit(oneString).cast(IntegerType).as(gf_trd_leg_id), lit(oneString).as(gf_trd_phase_id),
          lit(nullString).cast(StringType).as(gf_trd_leg_adj_eff_date), lit(nullString).cast(StringType).as(gf_trd_leg_adj_trmnt_date),
          lit(nullString).cast(StringType).as(gf_trd_phase_adj_eff_date), lit(nullString).cast(StringType).as(gf_trd_phase_adj_trmnt_date),
          abs(col(total_securities_nominal_amount)).as(gf_underlying_amount), abs(col(initial_amount)).as(gf_settl_amount),
          when(col(market_operation_type).isin(CD, OPR, CR), PAYER)
            .when(col(market_operation_type).isin(GET, VR, VD, GET), RECEIVER).as(gf_trd_party_payer_recv_type),
          col(isin_id).as(g_undly_asset_isin_id), lit(X).as(g_call_put_option_type), lit(nullString).cast(StringType).as(g_exercise_option_type),
          lit(nullString).cast(StringType).as(g_option_exercise_freq_type), lit(nullString).cast(StringType).as(gf_opt_adj_ex_dt_desc),
          col(market_operation_start_date).as(gf_am_opt_fst_ex_adj_date), lit(nullString).cast(StringType).as(gf_option_strike_amount),
          col(coupon_rate_type).as(g_interest_rate_type), dateCalendar(col(coupon_term_days_number)).as(g_interest_pymt_frequency_type),
          lit(nullString).cast(StringType).as(gf_pmt_dt_ref_typo_desc), lit(nullString).cast(StringType).as(gf_upfront_pmt_amount),
          coalesce(when(col(coupon_rate_type) === F, col(return_interest_rate_per).cast(DecimalType(fifteen, four))), lit(nullString)).as(gf_fixed_rate_per),
          lit(nullString).cast(StringType).as(g_amortization_frequency_type),
          coalesce(when(col(coupon_rate_type) === V, col(return_interest_rate_per).cast(DecimalType(fifteen, four))), lit(nullString)).as(gf_float_rt_ini_per),
          col(return_interest_rate_per).cast(DecimalType(fifteen, four)).as(gf_current_interest_rate_per),
          lit(nullString).cast(StringType).as(gf_net_life_cap_per), lit(nullString).cast(StringType).as(gf_net_life_floor_per),
          lit(nullString).cast(StringType).as(gf_swap_tenor_months_number), lit(nullString).cast(StringType).as(g_day_count_convention_type),
          dateCalendar(col(coupon_term_days_number)).as(g_compounding_frequency_type), lit(nullString).cast(StringType).as(g_renewal_frequency_type),
          concat((split(col(coupon_rate_name), "/")(1)), lit(" "), (split(col(coupon_rate_name), "/")(2))).as(g_reference_index_id),
          lit(nullString).cast(StringType).as(g_amortization_type),
          when(datediff(col(market_operation_settlement_date), date_format(lit(hdate), "yyyy-MM-dd")) > 2, col(current_commitment_amount))
            .otherwise(nullString).as(gf_forward_price_amount),
          lit(nullString).cast(StringType).as(gf_fx_forward_exch_rate_amount), lit(nullString).cast(StringType).as(gf_am_opt_lst_ex_adj_date),
          concat(
            lit(MX0074),
            when(col(seller_branch_id).isNull, lit(FourZeros))
              .otherwise(lpad(trim(col(seller_branch_id)), four, zero_string)),
            when(col(operation_deal_front_star_id) === emptyString, lit(ten_zero))
              .when(col(operation_deal_front_star_id) === zero_string, lit(ten_zero))
              .when(col(operation_deal_front_star_id) === 0, col(operation_deal_back_star_id))
              .otherwise(col(operation_deal_front_star_id))
          ).as(gf_local_contract_number_id),
          col(return_interest_rate_per).as(gf_sec_irr_per),
          col(guarantee_type).as(gf_colat_typo_desc),
          col(isin_id_coll).as(gf_collateral_isin_desc),
          col(current_guaranteed_amount).as(gf_colat_nmnl_amt_desc),
          col(return_interest_rate_per).cast(DecimalType(twelve, nine)).as(gf_initial_interest_rate_per),
          coalesce(catalogPrsPcv.col(value), lit(X)).as(g_trade_position_type),
          col(current_rate_coupon_per).cast(DecimalType(twelve, nine)).as(gf_ref_index_interest_rate_per)
        )
        .distinct()
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def forexTrading(sireFxPrdt: DataFrame, tsiOperacion: DataFrame, tsiFwdDiv: DataFrame, catalogApp: DataFrame,
                   catalogPrsPcv: DataFrame, guaranteeCore: DataFrame, catalogPrev: DataFrame): DataFrame = {
    try {
      sireFxPrdt.join(tsiOperacion, Seq(operation_deal_front_star_id, counterpart_id), left)
        .join(tsiFwdDiv, Seq(operation_deal_front_star_id), left)
        .join(catalogApp, sireFxPrdt.col(source_application_id) === catalogApp.col(id), left)
        .join(broadcast(catalogPrsPcv), sireFxPrdt.col(operation_nature_type) === catalogPrsPcv.col(id), left)
        .join(guaranteeCore, col(operation_deal_front_star_id) === guaranteeCore.col(contract_id), left)
        .join(catalogPrev, col(calculation_base_currency_type) === col(local_type), left)
        .select(coalesce(catalogApp.col(value), when(sireFxPrdt.col(source_application_id) === emptyString, BGM_BIS)
          .otherwise(sireFxPrdt.col(source_application_id))).as(gf_origin_application_id),
          coalesce(catalogApp.col(value), when(sireFxPrdt.col(source_application_id) === emptyString, BGM_BIS)
            .otherwise(sireFxPrdt.col(source_application_id))).as(gf_front_sys_name),
          col(operation_deal_front_star_id).as(gf_front_sys_trd_id), col(operation_deal_back_star_id).as(gf_back_sys_trd_id),
          lit(oneString).cast(IntegerType).as(gf_trd_leg_id), lit(oneString).as(gf_trd_phase_id),
          lit(nullString).cast(StringType).as(gf_trd_leg_adj_eff_date), lit(nullString).cast(StringType).as(gf_trd_leg_adj_trmnt_date),
          lit(nullString).cast(StringType).as(gf_trd_phase_adj_eff_date), lit(nullString).cast(StringType).as(gf_trd_phase_adj_trmnt_date),
          col(market_oper_recv_notional_amount).as(gf_underlying_amount), abs(col(market_oper_recv_notional_amount)).alias(gf_settl_amount),
          when(col(operation_nature_type) === COMPRA, PAYER)
            .when(col(operation_nature_type) === VENTA, RECEIVER).as(gf_trd_party_payer_recv_type),
          lit(emptyString).as(g_undly_asset_isin_id), lit(X).as(g_call_put_option_type),
          lit(nullString).cast(StringType).as(g_exercise_option_type), lit(nullString).cast(StringType).as(g_option_exercise_freq_type),
          lit(nullString).cast(StringType).as(gf_opt_adj_ex_dt_desc), lit(nullString).cast(StringType).as(gf_am_opt_fst_ex_adj_date),
          lit(nullString).cast(StringType).as(gf_option_strike_amount), lit(nullString).cast(StringType).as(g_interest_rate_type),
          lit(nullString).cast(StringType).as(g_interest_pymt_frequency_type), lit(nullString).cast(StringType).as(gf_pmt_dt_ref_typo_desc),
          lit(nullString).cast(StringType).as(gf_upfront_pmt_amount), lit(nullString).cast(StringType).as(gf_fixed_rate_per),
          lit(nullString).cast(StringType).as(g_amortization_frequency_type), lit(nullString).cast(StringType).as(gf_float_rt_ini_per),
          lit(nullString).cast(StringType).as(gf_current_interest_rate_per), lit(nullString).cast(StringType).as(gf_net_life_cap_per),
          lit(nullString).cast(StringType).as(gf_net_life_floor_per), lit(nullString).cast(StringType).as(gf_swap_tenor_months_number),
          when(col(holding_id).isNull, lit("XX")).otherwise(col(holding_id)).as(g_day_count_convention_type),
          lit(nullString).cast(StringType).as(g_compounding_frequency_type), lit(nullString).cast(StringType).as(g_renewal_frequency_type),
          lit(nullString).cast(StringType).as(g_reference_index_id), lit(nullString).cast(StringType).as(g_amortization_type),
          col(forward_price_amount).as(gf_forward_price_amount), col(forward_price_amount).as(gf_fx_forward_exch_rate_amount),
          lit(nullString).cast(StringType).as(gf_am_opt_lst_ex_adj_date),
          concat(
            lit(MX0074),
            when(col(seller_branch_id).isNull, lit(FourZeros))
              .otherwise(lpad(trim(col(seller_branch_id)), four, zero_string)),
            when(col(operation_deal_front_star_id) === emptyString, lit(ten_zero))
              .when(col(operation_deal_front_star_id) === zero_string, lit(ten_zero))
              .when(col(operation_deal_front_star_id) === 0, col(operation_deal_back_star_id))
              .otherwise(col(operation_deal_front_star_id))
          ).as(gf_local_contract_number_id),
          lit(nullString).cast(StringType).as(gf_sec_irr_per),
          col(guarantee_type).as(gf_colat_typo_desc), col(isin_id_coll).as(gf_collateral_isin_desc),
          col(current_guaranteed_amount).as(gf_colat_nmnl_amt_desc),
          lit(nullString).cast(DecimalType(twelve, nine)).as(gf_initial_interest_rate_per),
          coalesce(catalogPrsPcv.col(value), lit(X)).as(g_trade_position_type),
          lit(nullString).cast(DecimalType(twelve, nine)).as(gf_ref_index_interest_rate_per)
        )
        .distinct()
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }


  def organizedMarkets(derivadosMmooBis: DataFrame, tsiOperacion: DataFrame, tsiSwapsTasa: DataFrame, tsiOperacionMmo: DataFrame,
                       catalogOptionType: DataFrame, catalogApp: DataFrame, guaranteeCore: DataFrame,
                       catalogPrev: DataFrame): DataFrame = {
    try {
      derivadosMmooBis.join(tsiOperacion, Seq(operation_deal_front_star_id, counterpart_id), left)
        .join(tsiSwapsTasa, Seq(operation_deal_front_star_id), left)
        .join(catalogOptionType, col(warrant_type) === col(gf_catlg_field_value_es_desc), left)
        .join(tsiOperacionMmo, Seq(operation_deal_front_star_id, counterpart_id), left)
        .join(catalogApp, derivadosMmooBis.col(source_application_id) === catalogApp.col(id), left)
        .join(guaranteeCore, col(operation_deal_front_star_id) === guaranteeCore.col(contract_id), left)
        .join(catalogPrev, col(calculation_interest_name) === col(local_type), left)
        .select(coalesce(catalogApp.col(value), when(derivadosMmooBis.col(source_application_id) === emptyString, BGM_BIS)
          .otherwise(derivadosMmooBis.col(source_application_id))).as(gf_origin_application_id),
          coalesce(catalogApp.col(value), when(derivadosMmooBis.col(source_application_id) === emptyString, BGM_BIS)
            .otherwise(derivadosMmooBis.col(source_application_id))).as(gf_front_sys_name),
          col(operation_deal_front_star_id).as(gf_front_sys_trd_id), col(operation_deal_back_star_id).as(gf_back_sys_trd_id),
          coalesce(when(col(product_type) =!= F3, col(fin_derv_leg_exch_int_mark_type) + 1), lit(oneString)).cast(IntegerType).as(gf_trd_leg_id),
          coalesce(when(col(product_type) =!= F3, when(col(operation_stage_number) === 0, col(operation_stage_number) + 1).
            otherwise(col(operation_stage_number))), lit(oneString)).as(gf_trd_phase_id),
          coalesce(when(col(product_type) =!= F3, col(fin_derv_active_value_start_date)), lit(nullString)).cast(StringType).as(gf_trd_leg_adj_eff_date),
          coalesce(when(col(product_type) =!= F3, col(underlying_expiry_date)), lit(nullString)).cast(StringType).as(gf_trd_leg_adj_trmnt_date),
          coalesce(when(col(product_type) =!= F3, col(fin_derv_oper_phase_start_date)), lit(nullString)).cast(StringType).as(gf_trd_phase_adj_eff_date),
          coalesce(when(col(product_type) =!= F3, col(fin_derv_oper_phase_end_date)), lit(nullString)).cast(StringType).as(gf_trd_phase_adj_trmnt_date),
          abs(when(col(trading_mark_type) === V, col(market_oper_pay_notional_amount)).otherwise(col(market_oper_recv_notional_amount)))
            .as(gf_underlying_amount),
          when(col(product_type) === IRS, col(fin_derivative_settlement_amount)).otherwise(col(market_oper_pay_notional_amount)).as(gf_settl_amount),
          when((when(col(product_type).isin(IRS, IRC), col(market_operation_type)).otherwise(col(trading_mark_type))) === C, PAYER)
            .otherwise(RECEIVER).as(gf_trd_party_payer_recv_type),
          col(isin_id).as(g_undly_asset_isin_id), coalesce(col(gf_catalog_val_id), lit(X)).as(g_call_put_option_type),
          col(option_style_mark_type).as(g_exercise_option_type), lit(nullString).cast(StringType).as(g_option_exercise_freq_type),
          coalesce(when(col(product_type).isin(optProd: _*), col(expiration_date)), lit(nullString)).cast(StringType).as(gf_opt_adj_ex_dt_desc),
          coalesce(when(col(option_style_mark_type) === A, col(value_date)), lit(nullString)).cast(StringType).as(gf_am_opt_fst_ex_adj_date),
          derivadosMmooBis.col(strike_amount).as(gf_option_strike_amount),
          coalesce(when(col(product_type) =!= F3, (when(col(fin_derv_leg_exch_int_mark_type) === zero_string, F).
            when(col(fin_derv_leg_exch_int_mark_type) === oneString, V))), lit(nullString)).cast(StringType).as(g_interest_rate_type),
          coalesce(when(col(product_type) =!= F3, col(fin_derv_leg_flow_frequency_name)), lit(nullString)).cast(StringType).as(g_interest_pymt_frequency_type),
          lit(nullString).cast(StringType).as(gf_pmt_dt_ref_typo_desc), lit("-").as(gf_upfront_pmt_amount),
          coalesce(when(col(product_type) =!= F3, (when(col(fin_derv_leg_exch_int_mark_type) === zero_string, col(current_spread_amount)))), lit(nullString))
            .cast(StringType).as(gf_fixed_rate_per),
          coalesce(when(col(product_type) =!= F3, col(fin_derv_leg_flow_frequency_name)), lit(nullString)).as(g_amortization_frequency_type),
          coalesce(when(col(product_type) =!= F3, (when(col(fin_derv_leg_exch_int_mark_type) === oneString, col(interest_rate_amount)))), lit(nullString))
            .cast(StringType).as(gf_float_rt_ini_per),
          when(col(product_type) =!= F3, (when(col(fin_derv_leg_exch_int_mark_type) === oneString, col(forward_swap_pay_rate_amount))
            .otherwise(col(forward_swap_recv_rate_amount)))).as(gf_current_interest_rate_per),
          lit(nullString).cast(StringType).as(gf_net_life_cap_per), lit(nullString).cast(StringType).as(gf_net_life_floor_per),
          lit(nullString).cast(StringType).as(gf_swap_tenor_months_number),
          when(col(holding_id).isNull, lit("XX")).otherwise(col(holding_id)).as(g_day_count_convention_type),
          coalesce(when(col(product_type) =!= F3, col(fin_derv_leg_flow_frequency_name)), lit(nullString)).as(g_compounding_frequency_type),
          lit(nullString).cast(StringType).as(g_renewal_frequency_type),
          coalesce(when(col(product_type) =!= F3, when(col(fin_derv_leg_exch_int_mark_type) === oneString, col(underline_reference_type))), lit(emptyString))
            .cast(StringType).as(g_reference_index_id),
          coalesce(when(col(product_type) =!= F3, col(operation_repayment_number)), lit(nullString)).cast(StringType).as(g_amortization_type),
          col(contract_amount).as(gf_forward_price_amount), col(contract_amount).as(gf_fx_forward_exch_rate_amount),
          coalesce(when((col(product_type).isin(optProd: _*)) && (col(option_style_mark_type) === A), col(expiration_date)), lit(nullString))
            .cast(StringType).as(gf_am_opt_lst_ex_adj_date),
          concat(
            lit(MX0074),
            when(col(seller_branch_id).isNull, lit(FourZeros))
              .otherwise(lpad(trim(col(seller_branch_id)), four, zero_string)),
            when(col(operation_deal_front_star_id) === emptyString, lit(ten_zero))
              .when(col(operation_deal_front_star_id) === zero_string, lit(ten_zero))
              .when(col(operation_deal_front_star_id) === 0, col(operation_deal_back_star_id))
              .otherwise(col(operation_deal_front_star_id))
          ).as(gf_local_contract_number_id),
          lit(nullString).cast(StringType).as(gf_sec_irr_per), col(guarantee_type).as(gf_colat_typo_desc),
          col(isin_id_coll).as(gf_collateral_isin_desc),
          col(current_guaranteed_amount).as(gf_colat_nmnl_amt_desc),
          lit(nullString).cast(DecimalType(twelve, nine)).as(gf_initial_interest_rate_per),
          coalesce(when(col(product_type).isin(IRS, IRC),
            col(market_operation_type)).otherwise(col(trading_mark_type)), lit(X)).as(g_trade_position_type),
          lit(nullString).cast(DecimalType(twelve, nine)).as(gf_ref_index_interest_rate_per)
        )
        .distinct()
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }


  def overTheCounter(finDervOtcBis: DataFrame, activeOperations: DataFrame, catalogOptionType: DataFrame, subproductCfc: DataFrame,
                     catalogApp: DataFrame, tsiOperacionOtc: DataFrame, guaranteeCore: DataFrame, catalogPrev: DataFrame): DataFrame = {
    try {
      finDervOtcBis.join(activeOperations, Seq(operation_deal_front_star_id), left)
        .join(catalogOptionType, col(warrant_type) === col(gf_catlg_field_value_es_desc), left)
        .join(subproductCfc, Seq(operation_deal_front_star_id), left)
        .join(catalogApp, finDervOtcBis.col(source_application_id) === catalogApp.col(id), left)
        .join(tsiOperacionOtc, Seq(operation_deal_front_star_id, portfolio_type, counterpart_id), left)
        .join(guaranteeCore, col(operation_deal_front_star_id) === guaranteeCore.col(contract_id), left)
        .join(catalogPrev, col(coupon_rate_name) === col(local_type), left)
        .select(
          coalesce(catalogApp.col(value), when(finDervOtcBis.col(source_application_id) === emptyString, BGM_BIS)
            .otherwise(finDervOtcBis.col(source_application_id))).as(gf_origin_application_id),
          coalesce(catalogApp.col(value), when(finDervOtcBis.col(source_application_id) === emptyString, BGM_BIS)
            .otherwise(finDervOtcBis.col(source_application_id))).as(gf_front_sys_name),
          col(operation_deal_front_star_id).as(gf_front_sys_trd_id), col(operation_deal_back_star_id).as(gf_back_sys_trd_id),
          coalesce((col(swap_legs_id) + 1), (col(operation_parts_number) + 1), lit(oneString)).cast(IntegerType).as(gf_trd_leg_id),
          coalesce(when(col(operation_stage_number) === 0, (col(operation_stage_number) + 1)).otherwise(col(operation_stage_number)),
            lit(oneString)).as(gf_trd_phase_id), col(contract_register_date).as(gf_trd_leg_adj_eff_date),
          col(market_operation_expiry_date).as(gf_trd_leg_adj_trmnt_date), col(market_operation_start_date).as(gf_trd_phase_adj_eff_date),
          coalesce(col(fin_derv_oper_phase_end_date), col(underlying_expiry_date)).as(gf_trd_phase_adj_trmnt_date),
          abs(col(nominal_amount).cast(DecimalType(twenty_six, six))).as(gf_underlying_amount),
          when(col(product_type) === IRS, col(fin_derivative_settlement_amount)).otherwise(lit(nullString)).as(gf_settl_amount),
          when((when(tsiOperacionOtc.col(market_operation_type).isNotNull, tsiOperacionOtc.col(market_operation_type))
            .otherwise(col(trading_mark_type))) === C, PAYER).otherwise(RECEIVER).as(gf_trd_party_payer_recv_type),
          col(isin_id).as(g_undly_asset_isin_id), coalesce(col(gf_catalog_val_id), lit(X)).as(g_call_put_option_type),
          col(option_style_mark_type).substr(0, 1).as(g_exercise_option_type), lit(nullString).cast(StringType).as(g_option_exercise_freq_type),
          when(col(product_type).contains(OPCIONES), col(expiration_date)).as(gf_opt_adj_ex_dt_desc),
          when(col(product_type).contains(OPCIONES) && col(option_style_mark_type) === AMERICANA, col(value_date)).as(gf_am_opt_fst_ex_adj_date),
          col(strike_amount).as(gf_option_strike_amount),
          coalesce(when(subproductCfc.col(fixed_var_flow_leg_mark_type) === F || col(fin_derv_leg_exch_int_mark_type) === zero_string, F)
            .when(subproductCfc.col(fixed_var_flow_leg_mark_type) === V || col(fin_derv_leg_exch_int_mark_type) === oneString, V),
            lit(nullString)).as(g_interest_rate_type),
          col(fin_derv_leg_flow_frequency_name).cast(StringType).as(g_interest_pymt_frequency_type),
          col(operations_date_payment_type).cast(StringType).as(gf_pmt_dt_ref_typo_desc), lit(nullString).cast(StringType).as(gf_upfront_pmt_amount),
          abs((coalesce(when(subproductCfc.col(fixed_var_flow_leg_mark_type) === F ||
            (col(fin_derv_leg_exch_int_mark_type) === zero_string &&
              col(product_type) =!= WARRANTN), coalesce(col(current_spread_amount), col(forward_swap_recv_rate_amount))),
            lit(nullString))).cast(DecimalType(twelve, nine))).as(gf_fixed_rate_per), col(fin_derv_leg_flow_frequency_name)
            .cast(StringType).as(g_amortization_frequency_type),
          abs((coalesce(when(subproductCfc.col(fixed_var_flow_leg_mark_type) === V ||
            (col(fin_derv_leg_exch_int_mark_type) === oneString &&
              col(product_type) =!= WARRANTN), coalesce(col(interest_rate_amount), col(forward_swap_pay_rate_amount))),
            lit(nullString))).cast(DecimalType(twelve, nine))).as(gf_float_rt_ini_per), when(col(fin_derv_leg_exch_int_mark_type) === oneString,
            col(forward_swap_pay_rate_amount))
            .otherwise(col(forward_swap_recv_rate_amount)).as(gf_current_interest_rate_per), col(maximum_operation_rate_per)
            .cast(StringType).as(gf_net_life_cap_per),
          col(minimum_operation_rate_per).cast(StringType).as(gf_net_life_floor_per),
          when(col(subproduct_type) === OSW, months_between(col(fin_derv_oper_phase_end_date), col(fin_derv_oper_phase_start_date)))
            .cast(DecimalType(eight, four)).as(gf_swap_tenor_months_number),
          when(col(holding_id).isNull, lit(XX)).otherwise(col(holding_id)).as(g_day_count_convention_type),
          col(fin_derv_leg_flow_frequency_name).as(g_compounding_frequency_type),
          lit(nullString).as(g_renewal_frequency_type),
          coalesce(when(subproductCfc.col(fixed_var_flow_leg_mark_type) === V || col(fin_derv_leg_exch_int_mark_type) === oneString,
            col(underline_reference_type)), lit(emptyString)).as(g_reference_index_id),
          col(operation_repayment_number).as(g_amortization_type), col(forward_price_amount).as(gf_forward_price_amount),
          col(forward_price_amount).as(gf_fx_forward_exch_rate_amount),
          coalesce(when(col(product_type).contains(OPCIONES) && col(option_style_mark_type) === AMERICANA, col(expiration_date)),
            lit(nullString)).as(gf_am_opt_lst_ex_adj_date),
          concat(
            lit(MX0074),
            when(col(seller_branch_id).isNull, lit(FourZeros))
              .otherwise(lpad(trim(col(seller_branch_id)), four, zero_string)),
            when(col(operation_deal_front_star_id) === emptyString, lit(ten_zero))
              .when(col(operation_deal_front_star_id) === zero_string, lit(ten_zero))
              .when(col(operation_deal_front_star_id) === zero, col(operation_deal_back_star_id))
              .otherwise(col(operation_deal_front_star_id))
          ).as(gf_local_contract_number_id),
          lit(emptyString).as(gf_sec_irr_per), col(guarantee_type).as(gf_colat_typo_desc),
          col(isin_id_coll).as(gf_collateral_isin_desc),
          col(current_guaranteed_amount).as(gf_colat_nmnl_amt_desc),
          lit(nullString).cast(DecimalType(twelve, nine)).as(gf_initial_interest_rate_per),
          coalesce(col(trading_mark_type), lit(X)).as(g_trade_position_type),
          lit(nullString).cast(DecimalType(twelve, nine)).as(gf_ref_index_interest_rate_per)
        )
        .distinct()
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def deposits(internalOperMurexDepos: DataFrame, tsiOperacion: DataFrame, tsiDepositos: DataFrame, guranteeCore: DataFrame,
               catalogPrev: DataFrame, mdcoMoneyMktBis: DataFrame): DataFrame = {
    try {
      internalOperMurexDepos
        .join(tsiOperacion, Seq(operation_deal_front_star_id), left)
        .join(tsiDepositos, Seq(operation_deal_front_star_id, sire_id), left)
        .join(guranteeCore, Seq(contract_id), left)
        .join(catalogPrev, col(calculation_interest_name) === col(local_type), left)
        .join(mdcoMoneyMktBis, Seq(operation_deal_front_star_id), left)
        .select(lit(MUREX).as(gf_origin_application_id), lit(MUREX).as(gf_front_sys_name),
          col(operation_deal_front_star_id).as(gf_front_sys_trd_id), lit(nullString).cast(StringType).as(gf_back_sys_trd_id),
          lit(oneString).cast(IntegerType).as(gf_trd_leg_id),
          coalesce(when(tsiDepositos.col(operation_stage_number) === zero, tsiDepositos.col(operation_stage_number) + 1)
            .otherwise(tsiDepositos.col(operation_stage_number)), lit(oneString)).as(gf_trd_phase_id),
          lit(nullString).cast(StringType).as(gf_trd_leg_adj_eff_date), lit(nullString).cast(StringType).as(gf_trd_leg_adj_trmnt_date),
          lit(nullString).cast(StringType).as(gf_trd_phase_adj_eff_date), lit(nullString).cast(StringType).as(gf_trd_phase_adj_trmnt_date),
          abs(col(current_operation_nom_amount)).as(gf_underlying_amount), lit(nullString).cast(StringType).as(gf_settl_amount),
          when(col(transaction_type) === B, PAYER).when(col(transaction_type) === S, RECEIVER).as(gf_trd_party_payer_recv_type),
          lit(nullString).cast(StringType).as(g_undly_asset_isin_id), lit(X).as(g_call_put_option_type),
          lit(nullString).cast(StringType).as(g_exercise_option_type), lit(nullString).cast(StringType).as(g_option_exercise_freq_type),
          lit(nullString).cast(StringType).as(gf_opt_adj_ex_dt_desc), lit(nullString).cast(StringType).as(gf_am_opt_fst_ex_adj_date),
          lit(nullString).cast(StringType).as(gf_option_strike_amount), col(fin_derv_leg_exch_int_mark_type).as(g_interest_rate_type),
          tsiDepositos.col(fin_derv_leg_flow_frequency_name).as(g_interest_pymt_frequency_type),
          lit(nullString).cast(StringType).as(gf_pmt_dt_ref_typo_desc), lit(nullString).cast(StringType).as(gf_upfront_pmt_amount),
          coalesce(when(col(fin_derv_leg_exch_int_mark_type) === F, col(forward_rate_amount)), lit(nullString)).as(gf_fixed_rate_per),
          lit(nullString).cast(StringType).as(g_amortization_frequency_type),
          coalesce(when(col(fin_derv_leg_exch_int_mark_type) === V, col(forward_rate_amount)), lit(nullString)).as(gf_float_rt_ini_per),
          col(forward_rate_amount).as(gf_current_interest_rate_per),
          lit(nullString).cast(StringType).as(gf_net_life_cap_per), lit(nullString).cast(StringType).as(gf_net_life_floor_per),
          lit(nullString).cast(StringType).as(gf_swap_tenor_months_number),
          when(col(holding_id).isNull, lit(XX)).otherwise(col(holding_id)).as(g_day_count_convention_type),
          tsiDepositos.col(fin_derv_leg_flow_frequency_name).as(g_compounding_frequency_type),
          tsiDepositos.col(fin_derv_leg_flow_frequency_name).as(g_renewal_frequency_type),
          col(underlying_main_index_id).as(g_reference_index_id),
          coalesce(col(amortization1_id), col(amortization_id)).as(g_amortization_type), col(forward_price_amount).as(gf_forward_price_amount),
          lit(nullString).cast(StringType).as(gf_fx_forward_exch_rate_amount), lit(nullString).cast(StringType).as(gf_am_opt_lst_ex_adj_date),
          concat(
            lit(MX0074),
            when(col(seller_branch_id).isNull, lit(FourZeros))
              .otherwise(lpad(trim(col(seller_branch_id)), four, zero_string)),
            col(operation_deal_front_star_id)
          ).as(gf_local_contract_number_id),
          lit(nullString).cast(StringType).as(gf_sec_irr_per), col(guarantee_type).as(gf_colat_typo_desc), col(isin_id_coll).as(gf_collateral_isin_desc),
          col(current_guaranteed_amount).as(gf_colat_nmnl_amt_desc),
          lit(nullString).cast(DecimalType(twelve, nine)).as(gf_initial_interest_rate_per),
          coalesce(col(transaction_type), lit(X)).as(g_trade_position_type),
          lit(nullString).cast(DecimalType(twelve, nine)).as(gf_ref_index_interest_rate_per)
        )
        .distinct()
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

  def tradeProd(fixedRent: DataFrame, forexTrading: DataFrame, organizedMarkets: DataFrame, overTheCounter: DataFrame, deposits: DataFrame): DataFrame = {
    try {
      fixedRent.union(forexTrading).union(organizedMarkets).union(overTheCounter).union(deposits)
        .select(
          when(col(gf_origin_application_id) === emptyString, BGM_BIS).otherwise(col(gf_origin_application_id))
            .as(gf_origin_application_id_aux),
          col(gf_front_sys_name).as(gf_front_sys_name_aux),
          col(gf_front_sys_trd_id).as(gf_front_sys_trd_id_aux),
          col(gf_back_sys_trd_id).as(gf_back_sys_trd_id_aux),
          col(gf_trd_leg_id).as(gf_trd_leg_id_aux),
          col(gf_trd_phase_id).as(gf_trd_phase_id_aux),
          col(gf_trd_leg_adj_eff_date).as(gf_trd_leg_adj_eff_date_aux),
          col(gf_trd_leg_adj_trmnt_date).as(gf_trd_leg_adj_trmnt_date_aux),
          col(gf_trd_phase_adj_eff_date).as(gf_trd_phase_adj_eff_date_aux),
          col(gf_trd_phase_adj_trmnt_date).as(gf_trd_phase_adj_trmnt_date_aux),
          col(gf_underlying_amount).as(gf_underlying_amount_aux),
          col(gf_settl_amount).as(gf_settl_amount_aux),
          col(gf_trd_party_payer_recv_type).as(gf_trd_party_payer_recv_type_aux),
          col(g_undly_asset_isin_id).as(g_undly_asset_isin_id_aux),
          col(g_call_put_option_type).as(g_call_put_option_type_aux),
          col(g_exercise_option_type).as(g_exercise_option_type_aux),
          col(g_option_exercise_freq_type).as(g_option_exercise_freq_type_aux),
          col(gf_opt_adj_ex_dt_desc).as(gf_opt_adj_ex_dt_desc_aux),
          col(gf_am_opt_fst_ex_adj_date).as(gf_am_opt_fst_ex_adj_date_aux),
          col(gf_option_strike_amount).as(gf_option_strike_amount_aux),
          col(g_interest_rate_type).as(g_interest_rate_type_aux),
          col(g_interest_pymt_frequency_type).as(g_interest_pymt_frequency_type_aux),
          col(gf_pmt_dt_ref_typo_desc).as(gf_pmt_dt_ref_typo_desc_aux),
          col(gf_upfront_pmt_amount).as(gf_upfront_pmt_amount_aux),
          col(gf_fixed_rate_per).as(gf_fixed_rate_per_aux),
          col(g_amortization_frequency_type).as(g_amortization_frequency_type_aux),
          col(gf_float_rt_ini_per).as(gf_float_rt_ini_per_aux),
          col(gf_current_interest_rate_per).as(gf_current_interest_rate_per_aux),
          col(gf_net_life_cap_per).as(gf_net_life_cap_per_aux),
          col(gf_net_life_floor_per).as(gf_net_life_floor_per_aux),
          col(gf_swap_tenor_months_number).as(gf_swap_tenor_months_number_aux),
          col(g_day_count_convention_type).as(g_day_count_convention_type_aux),
          col(g_compounding_frequency_type).as(g_compounding_frequency_type_aux),
          col(g_renewal_frequency_type).as(g_renewal_frequency_type_aux),
          col(g_reference_index_id).as(g_reference_index_id_aux),
          col(g_amortization_type).as(g_amortization_type_aux),
          col(gf_forward_price_amount).as(gf_forward_price_amount_aux),
          col(gf_fx_forward_exch_rate_amount).as(gf_fx_forward_exch_rate_amount_aux),
          col(gf_am_opt_lst_ex_adj_date).as(gf_am_opt_lst_ex_adj_date_aux),
          col(gf_local_contract_number_id).as(gf_local_contract_number_id_aux),
          col(gf_sec_irr_per).as(gf_sec_irr_per_aux),
          col(gf_colat_typo_desc).as(gf_colat_typo_desc_aux),
          col(gf_collateral_isin_desc).as(gf_collateral_isin_desc_aux),
          col(gf_colat_nmnl_amt_desc).as(gf_colat_nmnl_amt_desc_aux),
          when(col(g_trade_position_type).isNotNull && col(g_trade_position_type) =!= lit(nullString), col(g_trade_position_type))
            .otherwise(lit(X))
            .as(g_trade_position_type_aux),
          col(gf_initial_interest_rate_per).as(gf_initial_interest_rate_per_aux),
          col(gf_ref_index_interest_rate_per).as(gf_ref_index_interest_rate_per_aux)
        )
    } catch {
      case e: Exception => ExceptionUtils.getStackTrace(e)
        throw e
    }
  }

}
```
